```python

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]


plt.figure(figsize=(6, 4))
plt.plot(x, y, marker='o', linestyle='-')


for i, (xi, yi) in enumerate(zip(x, y)):
    plt.annotate(f'({xi}, {yi})', (xi, yi), textcoords="offset points", xytext=(0, 10), ha='center')


plt.title('Annotations on Line Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')


plt.grid(True)

plt.show()
```

```python
data = np.random.randn(1000)

plt.hist(data, bins = 25, color = 'darkblue', edgecolor = 'yellow')
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Customized Histogram with Density Plot')
plt.show()
```

```python

data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 2

plt.hist(data1, bins=30, stacked = True, color=['cyan'], edgecolor='black')
plt.hist(data2, bins=30, color=['Purple'], edgecolor='black')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Stacked Histogram')

plt.legend(['Dataset 1', 'Dataset 2'])

plt.show()
```

```python
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]

fig, ax = plt.subplots()

ax.pie(data, labels = cars, shadow= True, explode=[0.1,0.2,0.1,0.3,0.1,0.1], wedgeprops = {'linewidth': 1, 'edgecolor': "green"})
ax.legend(bbox_to_anchor=(1, 0, 0.5, 1))
plt.title("Pie Chart split up")
plt.show()
```

```python
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]

explode = (0.1, 0.0, 0.2, 0.3, 0.0, 0.0)

colors = ("orange", "cyan", "brown",
          "grey", "indigo", "beige")

wp = {'linewidth': 1, 'edgecolor': "green"}

def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)


fig, ax = plt.subplots(figsize=(10, 7))
wedges, texts, autotexts = ax.pie(data,
                                  autopct=lambda pct: func(pct, data),
                                  explode=explode,
                                  labels=cars,
                                  shadow=True,
                                  colors=colors,
                                  startangle=90,
                                  wedgeprops=wp,
                                  textprops=dict(color="magenta"))

ax.legend(wedges, cars,
          title="Cars",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")

ax.set_title("Customized pie chart")

plt.show()
```
